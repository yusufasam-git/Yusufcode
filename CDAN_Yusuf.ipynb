# **tSNE visualization**

```python
import torch
import numpy as np
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

from models.feature_extractor import get_resnet50_feature_extractor
from models.dann import DANNClassifier
from datasets.office31 import Office31Dataset
from torchvision import transforms
from torch.utils.data import DataLoader

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Load models
F, feature_dim = get_resnet50_feature_extractor()
F.load_state_dict(torch.load("/content/cdan/results/cdan_20260207_094602/F_cdan.pth"))
F = F.to(device)
F.eval()

# Load data
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406],
                         [0.229, 0.224, 0.225])
])

src = Office31Dataset("/content/office31/amazon", transform)
tgt = Office31Dataset("/content/office31/webcam", transform)

src_loader = DataLoader(src, batch_size=32, shuffle=False)
tgt_loader = DataLoader(tgt, batch_size=32, shuffle=False)

def extract(loader, label):
    feats, labs = [], []
    with torch.no_grad():
        for x, _ in loader:
            x = x.to(device)
            f = F(x).view(x.size(0), -1)
            feats.append(f.cpu().numpy())
            labs.append(np.full(len(f), label))
    return np.concatenate(feats), np.concatenate(labs)

src_f, src_l = extract(src_loader, 0)
tgt_f, tgt_l = extract(tgt_loader, 1)

all_f = np.vstack([src_f, tgt_f])
all_l = np.hstack([src_l, tgt_l])

tsne = TSNE(n_components=2)
Z = tsne.fit_transform(all_f)

plt.figure(figsize=(8,6))
plt.scatter(Z[all_l==0,0], Z[all_l==0,1], s=5, label="Source")
plt.scatter(Z[all_l==1,0], Z[all_l==1,1], s=5, label="Target")
plt.legend()
plt.title("t-SNE Feature Alignment")

plt.savefig(f"{results_dir}/tsne.png")
plt.show()
```

# **Confusion Matrix**

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

from models.feature_extractor import get_resnet50_feature_extractor
from models.dann import DANNClassifier
from datasets.office31 import Office31Dataset
from torchvision import transforms
from torch.utils.data import DataLoader

import os

# -----------------------------
# 1. Load model
# -----------------------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

F, feature_dim = get_resnet50_feature_extractor()
C = DANNClassifier(feature_dim, 31)   # <-- change 31 to your num_classes

F.load_state_dict(torch.load("/content/cdan/results/cdan_20260207_094602/F_cdan.pth"))
C.load_state_dict(torch.load("/content/cdan/results/cdan_20260207_094602/C_cdan.pth"))

F = F.to(device).eval()
C = C.to(device).eval()

# -----------------------------
# 2. Load target dataset
# -----------------------------
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406],
                         [0.229, 0.224, 0.225])
])

tgt = Office31Dataset("/content/office31/webcam", transform)
tgt_loader = DataLoader(tgt, batch_size=32, shuffle=False)

# -----------------------------
# 3. Collect predictions
# -----------------------------
all_preds = []
all_labels = []

with torch.no_grad():
    for x, y in tgt_loader:
        x = x.to(device)
        y = y.to(device)

        f = F(x).view(x.size(0), -1)
        logits = C(f)
        preds = logits.argmax(dim=1)

        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(y.cpu().numpy())

all_preds = np.array(all_preds)
all_labels = np.array(all_labels)

# -----------------------------
# 4. Compute confusion matrix
# -----------------------------
cm = confusion_matrix(all_labels, all_preds)

plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("True")

# -----------------------------
# 5. Save BEFORE show (important!)
# -----------------------------
results_dir = "/content/cdan/results/cdan_20260207_094602"
plt.savefig(os.path.join(results_dir, "confusion_matrix.png"), dpi=300, bbox_inches='tight')
plt.show()
```

# **prediction visualization**

```python
import matplotlib.pyplot as plt
import torch

def unnormalize(img):
    # Ensure mean and std are on the same device as img
    mean = torch.tensor([0.485, 0.456, 0.406]).view(3,1,1).to(img.device)
    std = torch.tensor([0.229, 0.224, 0.225]).view(3,1,1).to(img.device)
    return img * std + mean

# Load classifier
C = DANNClassifier(feature_dim, 31)
C.load_state_dict(torch.load("/content/cdan/results/cdan_20260207_094602/C_cdan.pth"))
C = C.to(device)
C.eval()

images, labels = next(iter(tgt_loader))
images = images.to(device)

with torch.no_grad():
    feats = F(images).view(images.size(0), -1)
    preds = C(feats).argmax(dim=1).cpu()

plt.figure(figsize=(12,6))
for i in range(8):
    plt.subplot(2,4,i+1)
    img = unnormalize(images[i]).cpu().permute(1,2,0)
    plt.imshow(img)
    plt.title(f"Pred: {preds[i].item()}")
    plt.axis("off")
plt.show()
```
